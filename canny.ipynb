```
import numpy as np
import matplotlib.image as mpimage
import matplotlib.pyplot as plt
import cv2
```

```
image = cv2.imread("/content/sample_data/test.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap="gray")
```

```
kernel_size = 5
smooth_image = cv2.GaussianBlur(gray, (kernel_size, kernel_size), 0)
plt.imshow(smooth_image, cmap="gray")
```

```
min_threshold = 100
max_threshold = 200
edges = cv2.Canny(smooth_image, min_threshold, max_threshold)
plt.imshow(edges, cmap="gray")
```
