```
import numpy as np
import matplotlib.image as mpimage
import matplotlib.pyplot as plt

image = mpimage.imread('/content/sample_data/test.jpg')
plt.imshow(image)
```

```
(ySize, xSize, channels) = image.shape
colorSelect = np.copy(image)
rgb_threshold = [200, 200, 200]
thresholdIndexes = (image[:, :, 0] < rgb_threshold[0]) | (image[:, :, 1] < rgb_threshold[1]) | (image[:, :, 2] < rgb_threshold[2])
colorSelect[thresholdIndexes] = [0, 0, 0]
plt.imshow(colorSelect)
```
